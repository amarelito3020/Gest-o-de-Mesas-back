```python
#APS ALISSON HENRIQUE DE OLIVEIRA BARROS RA 09026389
import random
import math

# Cardápio
menu = {
    1: ("Hambúrguer", 15.00),
    2: ("Pizza", 30.00),
    3: ("Refrigerante", 5.00),
    4: ("Café", 4.00),
    5: ("Salada", 10.00)
}

# Sistema de pedidos (dicionário para armazenar os pedidos de cada mesa)
pedidos_mesas = {}

# Função para registrar um pedido
def registrar_pedido():
    mesa = int(input("Número da mesa: "))

    # Se a mesa não estiver no dicionário, adiciona
    if mesa not in pedidos_mesas:
        pedidos_mesas[mesa] = []

    print("\nCardápio:")
    for codigo, (nome, preco) in menu.items():
        print(f"{codigo}. {nome} - R$ {preco:.2f}")

    while True:
        item = int(input("\nDigite o código do item (ou 0 para voltar ao menu:"))
        if item == 0:
            break
        if item in menu:
            pedidos_mesas[mesa].append(menu[item])
            print(f"Item {menu[item][0]} adicionado à mesa {mesa}.")
        else:
            print("Código do item inválido.")

# Função para ver os pedidos de uma mesa específica
def ver_pedidos():
    mesa = int(input("Número da mesa: "))
    if mesa in pedidos_mesas and pedidos_mesas[mesa]:
        print(f"\nPedidos da mesa {mesa}:")
        for item in pedidos_mesas[mesa]:
            print(f"- {item[0]}: R$ {item[1]:.2f}")
    else:
        print("Nenhum pedido encontrado para essa mesa.")

# Função para calcular a conta de uma mesa
def calcular_conta():
    mesa = int(input("Número da mesa: "))
    if mesa in pedidos_mesas and pedidos_mesas[mesa]:
        total = sum(item[1] for item in pedidos_mesas[mesa])
        taxa_servico = total * 0.10
        total_com_taxa = total + taxa_servico
        print(f"\nTotal: R$ {total:.2f}")
        print(f"Taxa de serviço (10%): R$ {taxa_servico:.2f}")
        print(f"Total com taxa: R$ {total_com_taxa:.2f}")
    else:
        print("Nenhum pedido encontrado para essa mesa.")

# Função para dividir a conta entre várias pessoas
def dividir_conta():
    mesa = int(input("Número da mesa: "))
    if mesa in pedidos_mesas and pedidos_mesas[mesa]:
        num_pessoas = int(input("Número de pessoas para dividir: "))
        total = sum(item[1] for item in pedidos_mesas[mesa])
        taxa_servico = total * 0.10
        total_com_taxa = total + taxa_servico
        valor_por_pessoa = total_com_taxa / num_pessoas
        print(f"\nTotal: R$ {total:.2f}")
        print(f"Total com taxa de serviço: R$ {total_com_taxa:.2f}")
        print(f"Cada pessoa deve pagar: R$ {valor_por_pessoa:.2f}")
    else:
        print("Nenhum pedido encontrado para essa mesa.")

# Função para fechar uma mesa e limpar os pedidos
def fechar_mesa():
    mesa = int(input("Número da mesa: "))
    if mesa in pedidos_mesas:
        calcular_conta()  # Exibe o valor total antes de fechar a mesa
        del pedidos_mesas[mesa]
        print(f"A mesa {mesa} foi fechada.")
    else:
        print("Mesa não encontrada.")

# Função para adicionar um novo item ao cardápio
def adicionar_item_menu():
    print("\n--- Adicionar Novo Item ao Cardápio ---")

    # Solicita o código do novo item
    novo_codigo = int(input("Digite o código do novo item: "))

    # Verifica se o código já está em uso
    if novo_codigo in menu:
        print("Esse código já está em uso. Escolha outro código.")
        return

    # Solicita o nome e o preço do novo item
    novo_nome = input("Digite o nome do novo item: ")
    novo_preco = float(input("Digite o preço do novo item: R$ "))

    # Adiciona o novo item ao cardápio
    menu[novo_codigo] = (novo_nome, novo_preco)
    print(f"{novo_nome} adicionado ao cardápio com o código {novo_codigo}.")

# Função para remover um item do cardápio
def remover_item_menu():
    print("\n--- Remover Item do Cardápio ---")
    codigo = int(input("Digite o código do item a ser removido: "))

    if codigo in menu:
        item_removido = menu[codigo][0]  # Pega o nome do item
        del menu[codigo]  # Remove o item do cardápio
        print(f"{item_removido} foi removido do cardápio.")
    else:
        print("Código do item inválido. Nenhum item foi removido.")

# Função para exibir o menu e gerenciar a escolha do usuário
def menu_opcoes():
    while True:
        print("\n--- Menu de Opções ---")
        print("1. Registrar Pedido")
        print("2. Ver Pedidos da Mesa")
        print("3. Calcular Total da Conta")
        print("4. Dividir Conta")
        print("5. Fechar Mesa")
        print("6. Mesas Ocupadas")
        print("7. Adicionar Item ao Cardápio")
        print("8. Remover Item do Cardápio")  # Nova opção
        print("9. Sair")

        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            registrar_pedido()
        elif opcao == 2:
            ver_pedidos()
        elif opcao == 3:
            calcular_conta()
        elif opcao == 4:
            dividir_conta()
        elif opcao == 5:
            fechar_mesa()
        elif opcao == 6:
            print(f"Mesas ocupadas: {list(pedidos_mesas.keys())}")  # Mostra as mesas ocupadas
        elif opcao == 7:
            adicionar_item_menu()  # Chama a função para adicionar item
        elif opcao == 8:
            remover_item_menu()  # Chama a função para remover item
        elif opcao == 9:
            print("Encerrando o programa...")
            break
        else:
            print("Opção inválida, tente novamente.")

# Executa o menu
menu_opcoes()
```
